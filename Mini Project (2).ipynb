```python
import pandas as pd

orders=pd.read_csv("orders.csv")
customers=pd.read_csv("customers.csv")
products=pd.read_csv("products.csv")
print(orders)
print(customers)
print(products)
```

# Data Cleaning

```python
#1.1 : Using fillna to Fill missing Values
orders['Quantity']=orders['Quantity'].fillna(orders['Quantity'].median())

#1.2 : Replacing missing Age in the Customers table with the average age of all customers.
customers['Age'] = customers['Age'].fillna(customers['Age'].mean())

#1.3 : Droping Orders With Status Cancelled
orders = orders[orders['Order Status'] != 'Cancelled']

print("Order Data")
print(orders)
print("Customers Data")
print(customers)
print("Products Data")
print(products)
```

```python
# Fill missing Quantity in the Orders table with the median value.
orders['Quantity']=orders['Quantity'].fillna(orders['Quantity'].median())
print(orders)
```

```python
#. Replace missing Age in the Customers table with the average age of all customers.
customers['Age']=customers['Age'].fillna(customers['Age'].mean())
print(customers)
```

```python
#Drop orders with the status Cancelled.
drop=orders[orders['Order Status']!='Cancelled']
print(drop)
```

```python
#Ensure no duplicate rows exist in any dataset.
orders.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
```

# Data Merging and Manipulation

```python
#Merge Orders and Products using Product ID
orders_products = pd.merge(orders, products, on='Product ID')
print(orders_products)
```

```python
#Merge the result with Customers using Customer ID
merge_data = pd.merge(orders_products, customers, on='Customer ID')
print(merge_data)
```

```python
# Creating a new column Order Value as Quantity * Price Per Unit
merge_data['Order Value'] = merged_data['Quantity'] * merge_data['Price Per Unit']
print(merge_data)
```

```python
# Filter data for orders where Order Value > $500.
filter=merge_data[merge_data['Order Value']>500]
print(filter)
```

```python
# Calculating total Order Value for each country
total_order_value_each_country=merge_data.groupby('Country')['Order Value'].sum()
print(total_order_value_each_country)
```

```python
# Identifing the top 5 products with the highest total sales (Order Value)
total_sales=merge_data.groupby('Product Name')['Order Value'].sum()
top5=total_sales.sort_values(ascending=False)
print(top5.head(5))
```

# Data Visualization

```python
# importing libraries
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

# Creating a bar chart showing total sales by Country

total_sales_by_country=merge_data.groupby('Country')['Order Value'].sum()
print(total_sales_by_country)
total_sales_by_country.plot(kind='bar')
plt.title("Sales by Country")
plt.xlabel('Country')
plt.ylabel('Sales')
plt.show()
```

```python
#Using a line chart to visualize sales trends over time.

monthly_sales = merge_data.groupby('Join Date')['Order Value'].sum()
print(monthly_sales)
monthly_sales.plot(kind='line', color='green')
plt.title('Sales Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=90)
plt.show()
```

```python
# Using a heatmap to analyze correlations between Quantity, Price Per Unit, and Order Value.

corr_matrix=merge_data[['Quantity','Price Per Unit','Order Value']].corr()
sn.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title("correlations between Quantity, Price Per Unit, and Order Value")
plt.show()
```

```python
# pie chart showing the percentage contribution of each product category to total sales

each_product_total_sales=merge_data.groupby('Category')['Order Value'].sum()
print(each_product_total_sales)
each_product_total_sales.plot(kind='pie',autopct='%1.0f%%',startangle=90)
plt.title("Each product category to total sales")
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# Visualize stock levels for all products using a horizontal bar chart

products_stock_level=merge_data.groupby('Product Name')['Stock Quantity'].sum()
print(products_stock_level)
products_stock_level.plot(kind='barh')
plt.title("stock levels for all products")
plt.xlabel('Stock Quantity')
plt.ylabel('Products')
plt.show()
```
